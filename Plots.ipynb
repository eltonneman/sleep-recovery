```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
data1 = pd.read_csv('wellness_1.csv')
data2 = pd.read_csv('wellness_2.csv')
data3 = pd.read_csv('wellness_3.csv')
data4 = pd.read_csv('wellness_4.csv')
data5 = pd.read_csv('wellness_5.csv')
data6 = pd.read_csv('wellness_6.csv')
data7 = pd.read_csv('wellness_7.csv')
data8 = pd.read_csv('wellness_8.csv')
data9 = pd.read_csv('wellness_9.csv')
data10 = pd.read_csv('wellness_10.csv')
data11 = pd.read_csv('wellness_11.csv')
data12 = pd.read_csv('wellness_12.csv')
data13 = pd.read_csv('wellness_13.csv')
data14 = pd.read_csv('wellness_14.csv')
data15 = pd.read_csv('wellness_15.csv')
data16 = pd.read_csv('wellness_16.csv')
```

```python
fig, axes = plt.subplots(4, 4, figsize=(20, 10))
datasets = [data1, data2, data3, data4, data5, data6, data7, data8,
            data9, data10, data11, data12, data13, data14, data15, data16]

for i, df in enumerate(datasets):
    ax = axes[i//4, i%4]
    bins = range(int(df['sleep_quality'].min()), int(df['sleep_quality'].max()) + 2)
    ax.hist(df['sleep_quality'].dropna(), bins=bins, edgecolor='black', align='left', rwidth=0.6)
    ax.set_title(f'Data {i+1}')
    ax.set_xlabel('Sleep Quality')
    ax.set_ylabel('Frequency')
plt.suptitle("Historgram of Sleep Quality",fontsize=14, fontweight="bold", y=1)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(8, 2, figsize=(24,20))

datasets = [data1, data2, data3, data4, data5, data6, data7, data8,
            data9, data10, data11, data12, data13, data14, data15, data16]

for i, df in enumerate(datasets):
    ax = axes[i//2, i%2]
    df = df.sort_values('effective_time_frame')
    
    ax.plot(df['effective_time_frame'], df['sleep_quality'])
    ax.set_title(f'Data {i+1}')
    ax.set_xlabel('Time')

    ax.set_ylabel('Sleep Quality')
    #ax.tick_params(axis='x', rotation=45)
    ax.xaxis.set_major_locator(MaxNLocator(5))
plt.tight_layout()
plt.show()
```
